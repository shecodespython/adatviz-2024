# **BMI fit**

Készítsetek (a saját és a kaggle adatokra is):

* eloszlás (hisztogram) diagramokat:
  * magasság
  * testsúly
  * BMI
* összefüggés (scatter plot) diagramokat:
  * magasság vs. testsúly
  * magasság vs. BMI
  * testsúly vs. BMI

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.set_style("white")
sns.set_palette("tab20")
```

```python
sns.color_palette()
```

```python
palette = sns.color_palette()
```

```python
own = palette[8]
kaggle = palette[18]

fig, axs = plt.subplots(1, 2, figsize=(3.5, 2))

axs[0].add_patch(plt.Rectangle((0, 0), 1, 1, color=own))
axs[0].set_xticks([])
axs[0].set_yticks([])
axs[0].set_xlim(0, 1)
axs[0].set_ylim(0, 1)
axs[0].set_title("Saját Adatok")

axs[1].add_patch(plt.Rectangle((0, 0), 1, 1, color=kaggle))
axs[1].set_xticks([])
axs[1].set_yticks([])
axs[1].set_xlim(0, 1)
axs[1].set_ylim(0, 1)
axs[1].set_title("Kaggle Adatok")

plt.tight_layout()
plt.show()
```

```python
url_kaggle = "https://raw.githubusercontent.com/vlewir/adatvizualizacio/refs/heads/main/BMI_fit_kaggle.csv"
url_own = "https://raw.githubusercontent.com/vlewir/adatvizualizacio/refs/heads/main/BMI_fit_own.csv"

df_kaggle = pd.read_csv(url_kaggle)
df_own = pd.read_csv(url_own)
```

## **Saját adatok**

```python
df_own.head()
```

```python
df_own.columns
```

### **Magasság**

```python
df_own["Magassag (m)"].describe()
```

```python
fig, ax = plt.subplots()

sns.histplot(data=df_own, x="Magassag (m)", color=palette[8], ax=ax)

ax.set_xlabel("Magasság (m)")
ax.set_ylabel("Gyakoriság")
ax.set_title("Magasság Adatok Eloszlása")

ax.spines[["top", "right", "left", "bottom"]].set_visible(False)

plt.tight_layout()
plt.show()
```
